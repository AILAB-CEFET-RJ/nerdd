```python
import pandas as pd
import nltk
import json
```

```python
# Certifique-se de que você baixou os recursos necessários
nltk.download('punkt')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Carregar o arquivo CSV
caminho_arquivo_csv = "/content/drive/MyDrive/denuncias_testes.csv"
df = pd.read_csv(caminho_arquivo_csv, sep="@@")
```

```python
df
```

```python
# Supondo que a coluna com o texto seja chamada 'texto'
coluna_texto = 'denuncias'  # Substitua pelo nome correto da coluna

# Função para tokenizar o texto
def tokenizar_texto(texto):
    return nltk.word_tokenize(texto)

# Aplicar a tokenização na coluna e construir a estrutura JSON
resultado_json = []
```

```python
for index, row in df.iterrows():
    texto_tokenizado = tokenizar_texto(row[coluna_texto])
    entrada_json = {
        "tokenized_text": texto_tokenizado,
        "ner": []
    }
    resultado_json.append(entrada_json)
```

```python
# Caminho para salvar o arquivo JSON no Google Drive
caminho_arquivo_json = "/content/drive/MyDrive/resultado_testes.json"

# Salvar o resultado em um arquivo JSON
with open(caminho_arquivo_json, "w", encoding="utf-8") as arquivo_json:
    json.dump(resultado_json, arquivo_json, indent=2, ensure_ascii=False)

print("Arquivo JSON salvo com sucesso em:", caminho_arquivo_json)
```

```python
# Salvar o resultado em um arquivo JSON (opcional)
with open("/content/drive/MyDrive/resultado_tokenizado.json", "w", encoding="utf-8") as f:
    json.dump(resultado_json, f, indent=2, ensure_ascii=False)
```
