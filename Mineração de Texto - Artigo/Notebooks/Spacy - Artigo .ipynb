```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
def tokenize(text):
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = [token for token in tokens if re.search('[a-zA-Z]', token)]
    stems = [stemmer.stem(t) for t in filtered_tokens if len(t) > 0]
    return stems
```

```python
def stopwords():
    STOP_WORDS = ['em', 'em o', 'em a','ano',' ',':']
    return set(list(nltk.corpus.stopwords.words("portuguese")) + list(STOP_WORDS))

complete_stopwords = stopwords()
```

```python
def filter(token):
    if token.lower() in complete_stopwords:
       return True
    else:
      return False
```

```python
def lemma(documentacao):
  lemma_list = []
  for token in documentacao:
    if not filter(token.text.lower()):
      if not re.search(u'[^a-zA-ZáéíóúÁÉÍÓÚâêîôÂÊÎÔãõÃÕçÇ: ]', token.text.lower()):
        lemma_list.append(token.lemma_.lower())
  lemma_list = ' '.join(lemma_list)
  return lemma_list
```

```python
spacy_lemma = []
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_lemma.append(lemma(df['spacy_doc'][x]))
```

```python
df.loc[:,'spacy_lemma'] = spacy_lemma
```

```python
def NER_ORG(doc):
  list_ORG = []
  if doc.ents:
    for ent in doc.ents:
      if ent.label_ == 'ORG':
        list_ORG.append(ent.text)
  return list_ORG
```

```python
def NER_PERSON(doc):
  list_PERSON = []
  if doc.ents:
    for ent in doc.ents:
      if ent.label_ == 'PERSON':
        list_PERSON.append(ent.text)
  return list_PERSON
```

```python
def NER_LOC(doc):
  list_LOCATION = []
  if doc.ents:
    for ent in doc.ents:
      if ent.label_ == 'LOC':
        list_LOCATION.append(ent.text)
  return list_LOCATION
```

```python
spacy_org = []
spacy_loc = []
spacy_person = []
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_org.append(NER_ORG(df['spacy_doc'][x]))
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_loc.append(NER_LOC(df['spacy_doc'][x]))
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_person.append(NER_LOC(df['spacy_doc'][x]))
```

```python
df.loc[:, 'spacy_org'] = spacy_org
```
