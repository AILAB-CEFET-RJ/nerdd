```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.chunk import conlltags2tree, tree2conlltags

# Baixe os recursos necessários do NLTK
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
```

```python
# Função para realizar o reconhecimento de entidades nomeadas
def portuguese_ner(sentence):
  ent = []
  sent = []
    # Tokenização da frase
  tokens = word_tokenize(sentence)

    # Obtenha as partes do discurso (POS tags)
  pos_tags = pos_tag(tokens)

    # Reconhecimento de entidades nomeadas (NER)
  ne_tree = ne_chunk(pos_tags)

    # Converta a árvore para o formato IOB (Inside, Outside, Beginning)
  iob_tags = tree2conlltags(ne_tree)

    # Exiba as entidades nomeadas
  for word, pos, ne in iob_tags:
        if ne != 'O':
            print(f"{word};{ne}")
            ent.append(sentence+";"+word+";"+"ne")
  return ent
```
