```python
def NER_TIME(doc):
  list_TIME = []
  if doc.ents:
    for ent in doc.ents:
      print(ent)
      if ent.label_ == 'TIME':
        list_TIME.append(ent.text)
  return list_TIME
```

```python
def NER_PERSON(doc):
  list_PERSON = []
  if doc.ents:
    for ent in doc.ents:
      if ent.label_ == 'PERSON':
        list_PERSON.append(ent.text)
  return list_PERSON
```

```python
def NER_LOC(doc):
  list_LOCATION = []
  if doc.ents:
    for ent in doc.ents:
      if ent.label_ == 'LOC':
        list_LOCATION.append(ent.text)
  return list_LOCATION
```

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
result_path = '/content/drive/MyDrive/Mestrado/Dissertação/Artigo/Resultados/'
train_path = '/content/drive/MyDrive/Mestrado/Mineração de Texto/Trabalhos/Artigo/Dataset/'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(train_path + 'Spacy/Treino - TráficodeDrogasArmas.csv',sep=";")
```

```python
for x in range(0,len(df['relato'])):
  spacy_docs.append(nlp(df['relato'][x]))
```

```python
df.loc[:, 'spacy_doc'] = spacy_docs
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_time.append(NER_TIME(df['spacy_doc'][x]))
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_loc.append(NER_LOC(df['spacy_doc'][x]))
```

```python
for x in range(0,len(df['spacy_doc'])):
  spacy_person.append(NER_PERSON(df['spacy_doc'][x]))
```

```python
ents = []
ent_label = []
numeros = []
```

```python
for i in range(len(df['spacy_doc'])):
  text_to_test = df['spacy_doc'][i]
  doc = nlp(text_to_test)
  print(i)

  # Imprimir as entidades reconhecidas
  for ent in doc.ents:
      print(f"{ent.text};{ent.label_}")
      ents.append(ent.text)
      ent_label.append(ent.label_)
```

```python
df_ents = pd.DataFrame(ents,columns =['Entidades'])
df_type_ents = pd.DataFrame(ent_label,columns =['Label'])
```

```python
df = pd.concat([df_ents,df_type_ents],axis=1)
```
