```python
import openai
from openai import OpenAI

# Configure com suas credenciais de API do OpenAI
client = OpenAI(api_key="YOUR-TOKEN")


def obter_resposta(texto_entrada):
    # Solicitar uma resposta ao ChatGPT
    resposta = client.chat.completions.create(
        messages = [
            {
            "role": "user",
            "content": texto_entrada,
            }
        ],
        model="gpt-3.5-turbo",
    )
    return resposta
```

```python
df = pd.read_csv(train_path + 'Treino - TráficodeDrogasArmas.csv',sep=";")
```

```python
texto = []
loc = []
org = []
person = []
```

```python
for i in range(len(df['relato'])):
  # Exemplo de uso
  texto.append(df['relato'][i])
  entrada_do_usuario = "Identificar o Locais no texto: " + df['relato'][i]
  resposta_do_chatgpt = obter_resposta(entrada_do_usuario)

  print("Usuário:", entrada_do_usuario)
  print("ChatGPT:", resposta_do_chatgpt.choices[0].message.content)
  loc.append(resposta_do_chatgpt.choices[0].message.content)
```

```python
for i in range(len(df['relato'])):
  # Exemplo de uso
  texto.append(df['relato'][i])
  entrada_do_usuario = "Identificar as Organizações no texto: " + df['relato'][i]
  resposta_do_chatgpt = obter_resposta(entrada_do_usuario)

  print("Usuário:", entrada_do_usuario)
  print("ChatGPT:", resposta_do_chatgpt.choices[0].message.content)
  org.append(resposta_do_chatgpt.choices[0].message.content)
```

```python
for i in range(len(df['relato'])):
  # Exemplo de uso
  entrada_do_usuario = "Identificar as Pessoas no texto: " + df['relato'][i]
  resposta_do_chatgpt = obter_resposta(entrada_do_usuario)

  print("Usuário:", entrada_do_usuario)
  print("ChatGPT:", resposta_do_chatgpt.choices[0].message.content)
  person.append(resposta_do_chatgpt.choices[0].message.content)
```

```python
type(person)
```

```python
df_person = pd.DataFrame(person,columns =['Person'])
df_location = pd.DataFrame(loc,columns =['Location'])
df_org = pd.DataFrame(org,columns =['Org'])
df_text = pd.DataFrame(texto,columns =['Text'])
```

```python
df = pd.concat([df_text,df_location,df_org,df_person],axis=1)
```

```python
df.to_csv(train_path + 'resultado_ChatGPT.csv')
```
